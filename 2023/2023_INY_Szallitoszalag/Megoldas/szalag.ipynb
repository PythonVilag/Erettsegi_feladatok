# 1. Feladat

```python
adatok = []

with open(file="szallit.txt", mode="r") as forras_file:
  for sor_szam, sor in enumerate(forras_file):
    sor = sor.strip().split()
    if sor_szam == 0:
      szalaghossz = int(sor[0])
      lepesido = int(sor[1])
    else:
      adat = {}
      adat["felkerult"] = int(sor[0])
      adat["indulashelye"] = int(sor[1])
      adat["erkezeshelye"] = int(sor[2])
      adat["tomeg"] = int(sor[3])
      adatok.append(adat)
```

# 2. Feladat

```python
print("2. feladat")

sor_szam = int(input("Adja meg, melyik adatsorra kíváncsi! ")) - 1
indulashelye = adatok[sor_szam]["indulashelye"]
erkezeshelye = adatok[sor_szam]["erkezeshelye"]

print(f"Honnan: {indulashelye} Hova: {erkezeshelye}")
```

# 3. Feladat

```python
def tav(szalaghossz, indulashelye, erkezeshelye):
  if erkezeshelye >= indulashelye:
    return erkezeshelye - indulashelye
  else:
    return szalaghossz - indulashelye + erkezeshelye
```

# 4. Feladat

```python
print("4. feladat")

max_tavolsag = 0
for adat in adatok:
  tavolsag = tav(szalaghossz, adat["indulashelye"], adat["erkezeshelye"])
  if tavolsag > max_tavolsag:
    max_tavolsag = tavolsag

print(f"A legnagyobb távolság: {max_tavolsag}")


max_tavolsagu_szallitasok = []
for index, adat in enumerate(adatok):
  tavolsag = tav(szalaghossz, adat["indulashelye"], adat["erkezeshelye"])
  if tavolsag == max_tavolsag:
    max_tavolsagu_szallitasok.append(index)

max_tavolsagu_szallitasok = " ".join([str(sz + 1) for sz in max_tavolsagu_szallitasok])

print(f"A maximális távolságú szállítások sorszáma: {max_tavolsagu_szallitasok}")
```
