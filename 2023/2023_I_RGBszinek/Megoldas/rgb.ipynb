# 1. Feladat

```python
kep = []

with open(file='kep.txt', mode='r') as forras_file:
  for sor in forras_file:
    sor = sor.strip().split()
    sor_ertekek = []
    rgb_ertekek = []
    for index, szam in enumerate(sor):
      rgb_ertekek.append(int(szam))
      if index % 3 == 2:
        sor_ertekek.append(rgb_ertekek)
        rgb_ertekek = []
    kep.append(sor_ertekek)
```

# 2. Feladat

```python
print("2. feladat:")

print("Kérem egy képpont adatait!")
sor_szam = int(input("Sor:")) - 1
oszlop_szam = int(input("Oszlop:")) - 1

rgb_ertekek = kep[sor_szam][oszlop_szam]
print(f"A képpont színe RGB({rgb_ertekek[0]},{rgb_ertekek[1]},{rgb_ertekek[2]})")
```

# 3. Feladat

```python
print("3. feladat:")

vilagos_keppontok = 0
for sor in kep:
  for rgb_ertekek in sor:
    if sum(rgb_ertekek) > 600:
      vilagos_keppontok += 1

print(f"A világos képpontok száma: {vilagos_keppontok}")
```

# 4. Feladat

```python
print("4. feladat:")

legsotetebb_keppont_ertek = 255 * 3
for sor in kep:
  for rgb_ertekek in sor:
    if sum(rgb_ertekek) < legsotetebb_keppont_ertek:
      legsotetebb_keppont_ertek = sum(rgb_ertekek)

print(f"A legsötétebb pont RGB összege: {legsotetebb_keppont_ertek}")


legsotetebb_keppontok = []
for sor in kep:
  for rgb_ertekek in sor:
    if sum(rgb_ertekek) == legsotetebb_keppont_ertek:
      legsotetebb_keppontok.append(rgb_ertekek)
      
print(f"A legsötétebb pixelek színe:")
for legsotetebb_keppont in legsotetebb_keppontok:
  print(f"RGB({legsotetebb_keppont[0]},{legsotetebb_keppont[1]},{legsotetebb_keppont[2]})")
```

# 5. Feladat

```python
def harat(sor_szam, elteres_ertek):
  for index in range(len(kep[sor_szam]) - 1):
    if abs(kep[sor_szam][index][2] - kep[sor_szam][index + 1][2]) > elteres_ertek:
      return True
  return False
```

# 6. Feladat

```python
print("6. feladat:")

felhos_sorok = []

for sor_szam in range(len(kep)):
  if harat(sor_szam, 10):
    felhos_sorok.append(sor_szam)

print(f"A felhő legfelső sora: {felhos_sorok[0] + 1}")
print(f"A felhő legalsó sora: {felhos_sorok[-1] + 1}")
```
