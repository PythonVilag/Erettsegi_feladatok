# 1. Feladat

A szokásos beolvasáson kívül egy belső paramétert is felveszünk: az utazás érvényességét.

````python
with open('utasadat.txt', 'r') as bemenet:
    adat = list()
    for sor in bemenet:
        sorszam, datumido, kartyaszam, tipus, lejarat = sor.split()
        adat.append({'sorszam': int(sorszam),
                     'datumido': datumido,
                     'kartyaszam': kartyaszam,
                     'tipus': tipus,
                     'lejarat': lejarat,
                     'ervenyes': None})
````

# 2. Feladat

````python
print('2. feladat')
print(f'A buszra {len(adat)} utas akart felszallni.')
````

# 3. Feladat

Az utazás érvényességét ellenőrizzük minden utas esetén, és az 1. feladatnál definiált változóba elmentjük, hogy az 5. feladatban ne kelljen ismét foglalkozni vele.

````python
print('3. feladat')
ervenytelenek_szama = 0
for utas in adat:
    if utas['tipus'] == 'JGY':
        if utas['lejarat'] == '0':
            utas['ervenyes'] = False
            ervenytelenek_szama += 1
        else:
            utas['ervenyes'] = True
    else:
        if utas['lejarat'] < utas['datumido'][:8]:
            utas['ervenyes'] = False
            ervenytelenek_szama += 1
        else:
            utas['ervenyes'] = True
print(f'A buszra {ervenytelenek_szama} utas nem szallhatott fel.')
````

# 4. Feladat

A `felszallok[k]` értéke azt mutatja, hogy a `k`-adik megállóban hányan akartak felszállni.

A `list.index()` függvény megkeresi a megadott érték első előfordulását a listán belül, visszatérési értéke ennek az előfordulásnak az indexe (ha egyáltalán nem fordul elő, akkor `None`-t ad vissza)

````python
print('4. feladat')
felszallok = [0 for _ in range(30)]
for utas in adat:
    megallo = utas['sorszam']
    felszallok[megallo] += 1

max_felszallok = max(felszallok)
megallo = felszallok.index(max_felszallok)
print(f'A legtobb utas ({max_felszallok} fő) a {megallo}. megalloban akart felszallni.')
````

Az `index()` függvény helyett használhatjuk az alábbi for ciklust is:

```
for k, n in enumerate(felszallok):
    if n == max_felszallok:
        megallo = k
        break
```

A `break` nélkül nem az első, hanem az utolsó maximális megállót kapnánk vissza. `break` nélkül két módon is megkereshetjük az elsőt:
*   vagy visszafelé megyünk, hiszen hátulról az utolsó az első 

```
for k, n in enumerate(reversed(felszallok)))
    if n == max_felszallok:
        megallo = k
```
*   vagy `while` loop-pal valósitjuk meg ugyanazt a logikát, amit a `break` eredményez


```
talalt = False
k = 0
while not talalt:
    if felszallok[k] == max_felszallok:
        megallo = k
        talalt = True
    k += 1
```
Ebben az esetben a `while` loop meg fog állni, mivel a listában mindig benne van a max elem, de általában vigyázni kell, hogy nehogy végtelen ciklust programozzunk.

# 6. Feladat

````python
def napokszama(e1, h1, n1, e2, h2, n2):
    h1 = (h1 + 9) % 12
    e1 -= h1 % 10
    d1 = 365*e1 + e1%4 - e1%100 + e1%400 +(h1*306+5)%10 +n1-1
    h2 = (h2 + 9) % 12
    e2 -= h2 % 10
    d2 = 365*e2 + e2%4 - e2%100 + e2%400 +(h2*306+5)%10 +n2-1
    return d2-d1
````

# 7. Feladat

A dátum formátuma mindig **ééééhhnn-óópp**, tehát az első 4 karakter az évet, a következő 2 a hónapot, a következő kettő pedig a napot jelöli. `str` tipusból `int` tipusba egyszerűen az `int()` függvénnyel válthatunk. Hasonlóan a lejárati dátum is, csak a **-óópp** hiányzik a végéről.

A kiírásnál arra kell figyelni, hogy a feladat által kért **éééé-hh-nn** formátumot használjuk. Az f-string formázásánál a `:02` azt jelenti, hogy a kiírt szám ot balról feltölti nullákkal 2 számjegyig.

````python
with open('figyelmezetetes.txt', 'w') as kimenet:
    for utas in adat:
        if utas['ervenyes'] and utas['tipus'] != 'JGY':
            datum = utas['datumido']
            e1 = int(datum[0:4])
            h1 = int(datum[4:6])
            n1 = int(datum[6:8])
            lejarat = utas['lejarat']
            e2 = int(lejarat[0:4])
            h2 = int(lejarat[4:6])
            n2 = int(lejarat[6:8])
            if napokszama(e1, h1, n1, e2, h2, n2) <= 3:
                print(utas['kartyaszam'], f'{e2:04}-{h2:02}-{n2:02}', file=kimenet)
````

Az érvényesség vizsgálata helyett azt is megtehetnénk, hogy a jegyeket kiszűrjük, és a `napokszama()<3` mellett a `0<=napokszama()`-t is megköveteljük. (pythonban egybe is írhatjuk: `if 0 <= napoksama < 3:`)
