# 1. Feladat

````python
with open('beosztas.txt', 'r') as bemenet:
  adatok = list()
  for sorszam, sor in enumerate(bemenet):
    if sorszam % 4 == 0:
      ora = dict()
      ora['tanar'] = sor.strip()
    elif sorszam % 4 == 1:
      ora['targy'] = sor.strip()
    elif sorszam % 4 == 2:
      ora['osztaly'] = sor.strip()
    else:
      ora['szam'] = int(sor.strip())
      adatok.append(ora)
````

# 2. Feladat

````python
print('2. feladat')
print(f'A fájlban {len(adatok)} bejegyzés van.')
````

# 3. Feladat

````python
osszes_tanitasi_ora = 0
for ora in adatok:
  osszes_tanitasi_ora += ora['szam']
print(f'Az iskolaban a heti osszoraszam: {osszes_tanitasi_ora}')
````

Egysoros megoldásként az első 3 sor helyett az alábbit is használhatjuk:


1. `osszes_tanitasi_ora = sum(ora['szam'] for ora in adatok)`

a `sum()` argumentuma - `(ora['szam'] for ora in adatok)` - egy generátor függvény. Olyan, mintha az óraszámok listáját adnánk be a függvénynek; azzal a különbséggel, hogy kevesebb memóriát foglal. De akár listaként is megadhatjuk (csak ez lassabb):
2. `osszes_tanitasi_ora = sum([ora['szam'] for ora in adatok])`

# 4. Feladat

````python
tanar = input('Egy tanár neve= ')

tanitasi_ora = 0
for ora in adatok:
  if ora['tanar'] == tanar:
    tanitasi_ora += ora['szam']
print(f'A tanár heti óraszáma: {tanitasi_ora}')
````

Ha a beolvasáskor nem `str.strip()`-elünk, akkor az összehasonlítás az 5. sorban sosem lesz igaz, mivel az `ora['tanar']` vegen ott marad az új sor (`'\n'`)karakter.

Általában ez nem okoz bajt, mivel a legtöbb beolvasásnál `str.split()`-et használunk, ami szintén figyelmen kívül hagyja a sorzáró u.n. "whitespace" karaktereket (`\s, \t, \n,...`)

# 5. Feladat

````python
osztalyok = dict()
for ora in adatok:
  if ora['targy'] == 'osztalyfonoki':
    osztaly = ora['osztaly']
    of = ora['tanar']
    osztalyok[osztaly] = of

with open('of.txt', 'w') as kimeneti_file:
  for osztaly, of in osztalyok.items():
    print(f'{osztaly} - {of}', file=kimeneti_file)
````

Ebben a megoldásban a txt végén van egy új sor. Ha a feladat külön kéri, hogy ne legyen, akkor a `print(..., end='') `alakját kell használni az utolsó sornál.

Ha sorba akarjuk rendezni (NEM KELL), akkor itt van arra is egy megoldás:
Az osztály évfolyamát `int`-ként kezeljük, aszerint rendezzük az `osztalyok` `dictionary`-t. Ha nem `int`-ként kezelnénk, akkor a `sorted()` betűrendbe rendezné, azaz 10.a hamarabb kerülne, mint a 9.a, mivel `'1'<'9'`.

```python
with open('of_sorted.txt', 'w') as kimenet:
  osztalyok = dict()
  for ora in adat:
    if ora['targy'] == 'osztalyfonoki':
      szam, betu = ora['osztaly'].split('.')
      osztaly = int(szam), betu
      of = ora['tanar']
      osztalyok[osztaly] = of
  for (szam, betu), of in sorted(osztalyok.items()):
    print(f'{szam}.{betu} - {of}', file=kimenet)
```

* Az osztályt szétbontjuk az évfolyam sorszámára, és az osztály betűjelére. (5. sor)
* Az évfolyamot átváltjuk `int` típusra. (6.sor)
* Az osztályok `dict`-be a `szam, betu` változópárt rakjuk be az összevont szöveg helyett (6-8. sor)
* A második `for` ciklusban az osztalyokon sorrendben megyünk végig a `sorted()` függvény használatával (9. sor)
* A második `for` ciklusban ismét szétbontjuk a `szam, betu` változópárt (9. sor)
* A kiírásnál előállítjuk az eredeti szöveget a `szam, betu` változópárból (10. sor)

# 6. Feladat

````python
osztaly = input('Osztály= ')
targy = input('Tantárgy= ')
szamlalo = 0

for ora in adatok:
  if ora['osztaly'] == osztaly and ora['targy'] == targy:
    szamlalo += 1

if szamlalo != 1:
  print('Csoportbontásban tanulják.')
else:
  print('Osztályszinten tanulják.')
````

Minden programban probléma a hibás input kezelése. Érettségi feladatok kifejezetten megmondják, hogy hibás adatot nem kell kezelni, de inputok esetén a tesztelést segítheti, ha mégis gondolunk pár dologra.

# 7. Feladat

A feladat nehézsége az, hogy ugyanazt a tanárt ne számoljuk többször. Ennek egy módja, hogy `set`-ben tároljuk őket, mivel a `set` tulajdonsága, hogy ha ugyanazt újra belerakjuk, akkor is csak egyszer lesz benne.

````python
tanarok = list()
for ora in adatok:
  if ora['tanar'] not in tanarok:
    tanarok.append(ora['tanar'])
print(f'Az iskolában {len(tanarok)} tanár tanít.')
````


Ha nem találkoztál még a `set`-tel, a legegyszerűbb úgy megjegyezni, mint egy érték nélküli `dict`.
* Hogy valami benne van-e a `set`-ben, a `<var> in set` szintaxissal tesztelhetjük.
* Belerakni egy elemet a `set.add()` függvénnyel lehet (`list.append()` analógja)
* Több elemet egyszerre a `set.update()`-tel lehet (`list.extend()` analógja)

Két törlés is van:
* `set.discard()` kitöröl egy elemet, ha megtalálja, nem csinál semmit, ha nem.
* `set.remove()` kitörli az elemet ha megtalálja, hibát dob, ha nem.
