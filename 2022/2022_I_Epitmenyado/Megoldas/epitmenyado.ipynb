# 1. Feladat

Az adatok beolvasásához az enumerate() függvényt fogjuk használni, miven vannak bizonyos sorok, amely értékeit a többitől eltérő módon kell kezelni. (Első sor jelenesetben.)

Az adatokat egy dictionary-kből álló listába mentjük el.

```python
telek_adatok = list()
adosavok = dict()

with open(file='utca.txt', mode='r') as forras_file:
  for sorszam, sor in enumerate(forras_file):
    if sorszam == 0:
      savok = sor.strip().split()
      adosavok['A'] = int(savok[0])
      adosavok['B'] = int(savok[1])
      adosavok['C'] = int(savok[2])

    else:
      telek_adat = dict()
      ertekek = sor.strip().split()

      telek_adat['tulajdonos'] = int(ertekek[0])
      telek_adat['utca'] = ertekek[1]
      telek_adat['hazszam'] = ertekek[2]
      telek_adat['adosav'] = ertekek[3]
      telek_adat['alapterulet'] = int(ertekek[4])

      telek_adatok.append(telek_adat)
```

# 2. Feladat

```python
print(f'2. feladat. A mintában {len(telek_adatok)} telek szerepel.')
```

# 3. Feladat

```python
print('3. feladat')

vizsgalt_tulajdonos = int(input('Egy tulajdonos adószáma: '))

vizsgalt_tulajdonos_telkek = list()

for telek_adat in telek_adatok:
  if telek_adat['tulajdonos'] == vizsgalt_tulajdonos:
    vizsgalt_tulajdonos_telkek.append(f"{telek_adat['utca']} utca {telek_adat['hazszam']}")

if len(vizsgalt_tulajdonos_telkek) == 0:
  print('Nem szerepel az adatállományban.')
else:
  for vizsgalt_tulajdonos_telek in vizsgalt_tulajdonos_telkek:
    print(vizsgalt_tulajdonos_telek)
```

# 4. Feladat

```python
def ado(adosav, alapterulet):
  szorzo = adosavok[adosav]
  fizetendo_ado = szorzo * alapterulet
  
  return fizetendo_ado
```

# 5. Feladat

```python
sav_total = {'A': [0, 0], 'B': [0, 0], 'C': [0, 0]}

for telek_adat in telek_adatok:
  adosav = telek_adat['adosav']
  alapterulet = telek_adat['alapterulet']

  sav_total[adosav][0] += 1
  if ado(adosav, alapterulet) >= 10000:
    sav_total[adosav][1] += ado(adosav, alapterulet)

for sav in sav_total:
  print(f'{sav} sávba {sav_total[sav][0]} telek esik, az adó {sav_total[sav][1]} Ft.')
```

# 6. Feladat

```python
egyedi_utcak = list()

for telek_adat in telek_adatok:
  if telek_adat['utca'] not in egyedi_utcak:
    egyedi_utcak.append(telek_adat['utca'])

print("6. feladat. A több sávba sorolt utcák:")
for egyedi_utca in egyedi_utcak:
  egyedi_utca_savok = list()

  for telek_adat in telek_adatok:
    if telek_adat['utca'] == egyedi_utca:
      if telek_adat['adosav'] not in egyedi_utca_savok:
        egyedi_utca_savok.append(telek_adat['adosav'])

  if len(egyedi_utca_savok) != 1:
    print(egyedi_utca)
```

#7. Feladat

```python
egyedi_tulajdonosok = list()

for telek_adat in telek_adatok:
  if telek_adat['tulajdonos'] not in egyedi_tulajdonosok:
    egyedi_tulajdonosok.append(telek_adat['tulajdonos'])

with open(file='fizetendo.txt', mode='w') as kimeneti_file:
  for egyedi_tulajdonos in egyedi_tulajdonosok:
    fizetendo_sum = 0

    for telek_adat in telek_adatok:
      if telek_adat['tulajdonos'] == egyedi_tulajdonos:
        adosav = telek_adat['adosav']
        alapterulet = telek_adat['alapterulet']

        fizetendo = ado(adosav, alapterulet)
        if fizetendo >= 10000:
          fizetendo_sum += fizetendo


    print(f'{egyedi_tulajdonos} {fizetendo_sum}', file=kimeneti_file)
```
