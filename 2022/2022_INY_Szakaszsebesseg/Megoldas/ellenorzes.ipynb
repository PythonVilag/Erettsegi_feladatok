# 1. Feladat

```python
adatok = list()

with open(file='meresek.txt', mode='r') as forras_file:
  for sor in forras_file:
    adatok.append(sor.strip().split())
```

# 2. Feladat

```python
print('2. feladat')

print(f'A mérés során {len(adatok)} jármű adatait rögzítették.')
```

# 3. Feladat

```python
print('\n3. feladat')

athaladt_jarmu_9_elott = 0

for adat in adatok:
  if int(adat[5]) < 9:
    athaladt_jarmu_9_elott += 1

print(f'9 óra előtt {athaladt_jarmu_9_elott} jármű haladt el a végponti mérőnél.')
```

# 4. Feladat

```python
print('\n4. feladat')
bekert_ido = input('Adjon meg egy óra és perc értéket! ').split()
bekert_ora, bekert_perc = int(bekert_ido[0]), int(bekert_ido[1])

adott_percben_athaladt_jarmu = 0

for adat in adatok:
  if (int(adat[1]) == bekert_ora) and (int(adat[2]) == bekert_perc):
    adott_percben_athaladt_jarmu += 1

print(f'\ta. A kezdeti méréspontnál elhaladt járművek száma: {adott_percben_athaladt_jarmu}')


utszakaszon_levo_jarmu = 0

for adat in adatok:
  if (int(adat[2]) < bekert_perc + 1) and (int(adat[6]) >= bekert_perc):
    utszakaszon_levo_jarmu += 1
print(f'\tb. A forgalomsűrűség: {round(utszakaszon_levo_jarmu / 10, 1)}')
```

# 5. Feladat

```python
print('\n5. feladat')

elso_adat_be = int(adatok[0][1]) + int(adatok[0][2])/60 + int(adatok[0][3])/3600 + int(adatok[0][4])/3600000
elso_adat_ki = int(adatok[0][5]) + int(adatok[0][6])/60 + int(adatok[0][7])/3600 + int(adatok[0][8])/3600000
szakasz_ido_min = elso_adat_ki - elso_adat_be

for adat in adatok:
  szakasz_be = int(adat[1]) + int(adat[2])/60 + int(adat[3])/3600 + int(adat[4])/3600000
  szakasz_ki = int(adat[5]) + int(adat[6])/60 + int(adat[7])/3600 + int(adat[8])/3600000

  szakasz_ido = szakasz_ki - szakasz_be

  if szakasz_ido < szakasz_ido_min:
    szakasz_ido_min = szakasz_ido
    leggyorsabb_rendszam = adat[0]
    leggyorsabb_szakasz_be = szakasz_be
    leggyorsabb_szakasz_ki = szakasz_ki
    leggyorsabb_sebesseg = 10 / (szakasz_ido)


print('A legnagyobb sebességgel haladó jármű')
print(f'\trendszáma: {leggyorsabb_rendszam}')
print(f'\tátlagsebessége: {int(leggyorsabb_sebesseg)} km/h')


leghagyott_jarmuvek = 0
for adat in adatok:
  szakasz_be = int(adat[1]) + int(adat[2])/60 + int(adat[3])/3600 + int(adat[4])/3600000
  szakasz_ki = int(adat[5]) + int(adat[6])/60 + int(adat[7])/3600 + int(adat[8])/3600000

  if (szakasz_be <= leggyorsabb_szakasz_be) and (szakasz_ki > leggyorsabb_szakasz_ki):
    leghagyott_jarmuvek += 1

print(f'\táltal lehagyott járművek száma: {leghagyott_jarmuvek}')
```

# 6. Feladat

```python
print('\n6. feladat')


gyorshajtok = 0

for adat in adatok:
  szakasz_be = int(adat[1]) + int(adat[2])/60 + int(adat[3])/3600 + int(adat[4])/3600000
  szakasz_ki = int(adat[5]) + int(adat[6])/60 + int(adat[7])/3600 + int(adat[8])/3600000

  szakasz_ido = szakasz_ki - szakasz_be

  if (10 / (szakasz_ido)) > 90:
    gyorshajtok += 1

print(f'A járművek {round((gyorshajtok / len(adatok))*100, 2)}%-a volt gyorshajtó.')
```

# 7. Feladat

```python
with open(file='buntetes.txt', mode='w') as kimeneti_file:
  for adat in adatok:
    szakasz_be = int(adat[1]) + int(adat[2])/60 + int(adat[3])/3600 + int(adat[4])/3600000
    szakasz_ki = int(adat[5]) + int(adat[6])/60 + int(adat[7])/3600 + int(adat[8])/3600000

    szakasz_ido = szakasz_ki - szakasz_be
    sebesseg = (10 / (szakasz_ido))
    buntetes = 0

    if 104 < sebesseg <= 121:
      buntetes = 30000
    elif 121 < sebesseg <= 136:
      buntetes = 45000
    elif 136 < sebesseg <= 151:
      buntetes = 60000
    elif 151 < sebesseg:
      buntetes = 200000

    if buntetes != 0:
      print(f'{adat[0]}\t {int(round(sebesseg, 0))} km/h\t{buntetes} Ft', file=kimeneti_file)

print('A fájl elkészült.')
```
