# 1. Feladat

```python
jelek = list()

with open(file='jel.txt', mode='r') as forras_file:
  for sor in forras_file:
    jel = dict()
    ertekek = sor.strip().split()

    jel['ora'] = int(ertekek[0])
    jel['perc'] = int(ertekek[1])
    jel['mp'] = int(ertekek[2])

    jel['x'] = int(ertekek[3])
    jel['y'] = int(ertekek[4])

    jelek.append(jel)
```

# 2. Feladat

```python
print('2. feladat')

sorszam = int(input('Adja meg a jel sorszámát! ')) - 1
print(f'x={jelek[sorszam]["x"]} y={jelek[sorszam]["y"]}')
```

# 3. Feladat

```python
def eletelt(ora_1, perc_1, mp_1, ora_2, perc_2, mp_2):
  eltelt_mp = 0

  eltelt_mp += (mp_2 - mp_1)
  eltelt_mp += (perc_2 - perc_1) * 60
  eltelt_mp += (ora_2 - ora_1) * 3600

  return eltelt_mp
```

# 4. Feladat

```python
print('4. feladat')

ora_1 = jelek[0]['ora']
perc_1 = jelek[0]['perc']
mp_1 = jelek[0]['mp']

ora_2 = jelek[-1]['ora']
perc_2 = jelek[-1]['perc']
mp_2 = jelek[-1]['mp']

eltelt_mp = eletelt(ora_1, perc_1, mp_1, ora_2, perc_2, mp_2)

eltel_ora = eltelt_mp // 3600
eltel_perc = (eltelt_mp - eltel_ora*3600)  // 60
eltelt_mp = eltelt_mp - (eltel_ora*3600 + eltel_perc*60)

print(f'Időtartam: {eltel_ora}:{eltel_perc}:{eltelt_mp}')
```

# 5. Feladat

```python
print('5. feladat')

x_koordinatak = [jel['x'] for jel in jelek]
y_koordinatak = [jel['y'] for jel in jelek]

x_max = max(x_koordinatak)
x_min = min(x_koordinatak)
y_max = max(y_koordinatak)
y_min = min(y_koordinatak)

print(f'Bal alsó: {x_min} {y_min}, jobb felső: {x_max} {y_max}')
```

# 6. Feladat

```python
print('6. feladat')

elmuzdulas = 0

for i in range(len(jelek) - 1):
  x_1 = jelek[i]['x']
  y_1 = jelek[i]['y']
  x_2 = jelek[i + 1]['x']
  y_2 = jelek[i + 1]['y']

  elmuzdulas += ((x_1 - x_2)**2 + (y_1 - y_2)**2)**(1/2)


print(f'Elmozdulás: {round(elmuzdulas, 3)} egység')
```

# 7. Feladat

```python
with open('kimaradt.txt', 'w') as kimeneti_file:
  for i in range(len(jelek) - 1):
  
    ora_1 = jelek[i]['ora']
    perc_1 = jelek[i]['perc']
    mp_1 = jelek[i]['mp']
    x_1 = jelek[i]['x']
    y_1 = jelek[i]['y']

    ora_2 = jelek[i + 1]['ora']
    perc_2 = jelek[i + 1]['perc']
    mp_2 = jelek[i + 1]['mp']
    x_2 = jelek[i + 1]['x']
    y_2 = jelek[i + 1]['y']

    # Koordináta eltérés
    koordinata_elteres = max((abs(x_2 - x_1) - 1) // 10, (abs(y_2 - y_1) - 1) // 10)

    # Idő eltérés
    eltelt_mp = eletelt(ora_1, perc_1, mp_1, ora_2, perc_2, mp_2)
    ido_elteres = (eltelt_mp - 1) // (5 * 60)

    if (ido_elteres > 0) or (koordinata_elteres > 0):
      if koordinata_elteres >= ido_elteres:
        print(f"{ora_2} {perc_2} {mp_2} koordináta-eltérés {int(koordinata_elteres)}", file=kimeneti_file)
      else:
        print(f"{ora_2} {perc_2} {mp_2} időeltérés {int(ido_elteres)}", file=kimeneti_file)
```
