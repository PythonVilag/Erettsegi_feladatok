# Szükséges adatfájlok letöltése a futtató környezetbe

```python
!wget https://raw.githubusercontent.com/PythonVilag/Erettsegi_feladatok/main/2018/2018_INY_Fogadoora/Forrasok/fogado.txt
```

# 1. Feladat

```python
adatok = []

with open("fogado.txt", mode="r") as forras_file:
  for sor in forras_file:
    sor = sor.strip()
    sor = sor.split(" ")

    adat = {
        "nev": f"{sor[0]} {sor[1]}",
        "lefoglalt_idopont": sor[2],
        "foglalas_datum": sor[3]
    }
    adatok.append(adat)
```

# 2. Feladat

```python
print("2. feladat")

print(f"Foglalások száma: {len(adatok)}")
```

# 3. Feladat

```python
print("3. feladat")

bekert_nev = input("Adjon meg egy nevet: ")
foglalas_szam = len([adat["nev"] for adat in adatok if adat["nev"] == bekert_nev])

if foglalas_szam:
  print(f"{bekert_nev} néven {foglalas_szam} időpontfoglalás van.")
else:
  print("A megadott néven nincs időpontfoglalás.")
```
