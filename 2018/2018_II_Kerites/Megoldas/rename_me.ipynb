# 1. Feladat

````python
print('1. feladat')

bemeneti_filenev = input('Adja meg a bemeneti f√°jl nev√©t! ')
bemeneti_sor = int(input('Adja meg egy sor sz√°m√°t! ')) - 1
bemeneti_oszlop = int(input('Adja meg egy oszlop sz√°m√°t! ')) - 1
````

# 2. Feladat

````python
tablazat = list()
lepesek = list()
x = int()

with open(file=bemeneti_filenev, mode='r') as forras_file:
    for sorszam, sor in enumerate(forras_file):
        sor_list = sor.strip().split()
        sor_list = [int(elem) for elem in sor_list]
        if sorszam < 9:
            tablazat.append(sor_list)
        else:
            lepesek.append(sor_list)
````

# 3. Feladat

````python
print('\n3. feladat')

cella = tablazat[bemeneti_sor][bemeneti_oszlop]
def resztablazat(sor, oszlop):
    """
    A resztablazat szama jobbra 1-gyel n√∂vekszik 3 cell√°nk√©nt,
    lefel√© peding 3-mal. A bal fels≈ë pedig az 1-es szam√∫.
    """
    resztabla_sorszam = (sor//3)*3 + (oszlop//3) + 1
    return resztabla_sorszam

if cella == 0:
    print('Az adott helyet m√©g nem t√∂lt√∂tt√©k ki.')
else:
    print(f'Az adott helyen szerepl≈ë sz√°m: {cella}')

resztabla_sorszam = resztablazat(bemeneti_sor, bemeneti_oszlop)
print(f'A hely a(z) {resztabla_sorszam} r√©szt√°bl√°zathoz tartozik.')
````

# 4. Feladat

````python
print('\n4. feladat')

nullak_szama = 0
for sor_list in tablazat:
    nullak_szama += sor_list.count(0)
print(f'Az √ºres helyek ar√°nya: {round(100*nullak_szama/81, 1)}%')
````

# 5. Feladat

````python
print('\n5. feladat')

for lepes in lepesek:
    szam = lepes[0]
    sor = lepes[1] - 1
    oszlop = lepes[2] - 1

    print(f'A kiv√°lasztott sor: {sor+1} oszlop: {oszlop+1} a sz√°m: {szam}')
    if tablazat[sor][oszlop] != 0:
        print('A helyet m√°r kit√∂lt√∂tt√©k.')
    elif szam in tablazat[sor]: 
        print('Az adott sorban m√°r szerepel a sz√°m.')
    elif szam in [sor_list[oszlop] for sor_list in tablazat]:
        print('Az adott oszlopban m√°r szerepel a sz√°m.')
    else:
        resztabla_flag = False
        balfelso_sor = (sor//3)*3
        balfelso_oszlop = (oszlop//3)*3
        for sorszam in range(balfelso_sor, balfelso_sor+3):
            for oszlopszam in range(balfelso_oszlop, balfelso_oszlop+3):
                if szam == tablazat[sorszam][oszlopszam]:
                    resztabla_flag = True
        if resztabla_flag:
            print('Az adott r√©szt√°bl√°zatban m√°r szerepel a sz√°m.')
        else:
            print('A l√©p√©s megtehet≈ë.')
    print()
````
