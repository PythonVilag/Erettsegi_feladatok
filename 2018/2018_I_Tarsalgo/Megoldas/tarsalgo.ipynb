# 1. Feladat

```python
adatok = []

with open("ajto.txt", mode="r") as forras_file:
  for sor in forras_file:
    sor = sor.strip()
    sor = sor.split(" ")

    adat = {
        "ora": int(sor[0]),
        "perc": int(sor[1]),
        "idopont": int(sor[0]) * 60 + int(sor[1]),
        "szemely": int(sor[2]),
        "irany": sor[3]
    }
    adatok.append(adat)
```

# 2. Feladat

```python
print("2. feladat")

elso_belepo = adatok[0]["szemely"]
utolso_kilepo = reversed([adat for adat in adatok if adat["irany"] == "ki"])[0]

print(f"Az első belépő: {elso_erkezo}")
print(f"Az utolsó kilépő: {utolso_kilepo}")
```

# 3. Feladat

```python
szemelyek = sorted(set([adat["szemely"] for adat in adatok]))

with open("athaladas.txt", mode="r") as kimeneti_file:
  for szemely in szemelyek:
    athaladasok_szama = [adat["szemely"] for adat in adatok].count(szemely)
    print(f"{szemely} {athaladasok_szama}", file=kimeneti_file)
```

# 4. Feladat

```python
print("4. feladat")

tarsalgoban_maradt = []

for szemely in szemelyek:
  bennt = 0
  for adat in adatok:
    if adat["szemely"] == szemely:
      if adat["irany"] == "be":
        bennt += 1
      elif adat["irany"] == "ki":
        bennt -= 1

  if bennt:
    tarsalgoban_maradt.append(szemely)

print(f"A végén a társalgóban voltak: {' '.join(tarsalgoban_maradt)}")
```

# 5. Feladat

```python
print("5. feladat")

tarsalgo_letszam = 0
max_tarsalgo_letszam = 0
max_tarsalgo_ora = adatok[0]["ora"]
max_tarsalgo_perc = adatok[0]["perc"]

for adat in adatok:
  if adat["irany"] == "be":
    tarsalgo_letszam += 1
  elif adat["irany"] == "ki":
    tarsalgo_letszam -= 1

  if tarsalgo_letszam > max_tarsalgo_letszam:
    max_tarsalgo_letszam = tarsalgo_letszam
    max_tarsalgo_ora = adat["ora"]
    max_tarsalgo_perc = adat["perc"]

max_tarsalgo_idopont = f"{max_tarsalgo_ora}:{max_tarsalgo_perc}"

print(f"Például {max_tarsalgo_idopont}-kor voltak a legtöbben a társalgóban.")
```

# 7. Feladat

```python
print("7. feladat")


for adat in adatok:
  if adat["szemely"] == bekert_szemely:
    idopont = f"{adat['ora']}:{adat['perc']}"
    if adat["irany"] == "be":
      print(f"{idopont}-", end="")
    elif adat["irany"] == "ki":
      print(f"{idopont}")
```

# 8. Feladat

```python
print("8. feladat")

ossz_bentlet = 0
bennt_maradt = False


for adat in adatok:
  if adat["szemely"] == bekert_szemely:
    if adat["irany"] == "be":
      bemeneti_idopont = adat["idopont"]
      bennt_maradt = True
    elif adat["irany"] == "ki":
      kimeneti_idopont = adat["idopont"]
      ossz_bentlet += kimeneti_idopont - bemeneti_idopont
      bennt_maradt = False

if bennt_maradt:
  print(f"A(z) {bekert_szemely}. személy összesen {ossz_bentlet} percet volt bent, a megfigyelés végén a társalgóban volt.")
else:
  ossz_bentlet += 15 * 16 - bemeneti_idopont
  print(f"A(z) {bekert_szemely}. személy összesen {ossz_bentlet} percet volt bent, a megfigyelés végén nem volt a társalgóban.")
```
