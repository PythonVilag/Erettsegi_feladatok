# 1. Feladat

```python
adatok = list()

with open(file='vonat.txt', mode='r') as forras_file:
  for sor in forras_file:
    vonat_adat = sor.strip().split()
    adat = dict()
    adat["vonat"] = int(vonat_adat[0])
    adat["allomas"] = int(vonat_adat[1])
    adat["ora"] = int(vonat_adat[2])
    adat["perc"] = int(vonat_adat[3])
    adat["indul"] = vonat_adat[4]

    adatok.append(adat)
```

# 2. Feladat

```python
print('2. feladat')

allomasok = list()
vonatok = list()

for adat in adatok:
  if adat["allomas"] not in allomasok:
    allomasok.append(adat["allomas"])
  if adat["vonat"] not in vonatok:
    vonatok.append(adat["vonat"])

print(f'Az állomások száma:{len(allomasok)}')
print(f'A vonatok száma: {len(vonatok)}')
```

# 3. Feladat

```python
print('\n3. feladat')

max_allomas = adatok[0]["allomas"]
max_vonat = adatok[0]["vonat"]
max_varakozas = -1

for allomas in allomasok[1:-1]:
  for vonat in vonatok:
    for adat in adatok:
      if (adat["allomas"] == allomas) and (adat["vonat"] == vonat):
        if adat["indul"] == "I":
          indulasi_ido = adat["ora"] * 60 + adat["perc"]
        if adat['indul'] == "E":
          erkezesi_ido = adat["ora"] * 60 + adat["perc"]

    varakozas = indulasi_ido - erkezesi_ido
    if varakozas > max_varakozas:
      max_varakozas = varakozas
      max_allomas = allomas
      max_vonat = vonat

print(f'A(z) {max_vonat}. vonat a(z) {max_allomas}. állomáson {max_varakozas} percet állt.')
```

# 4. Feladat

```python
print('\n4. feladat')

bemeneti_vonat = int(input('Adja meg egy vonat azonosítóját! '))
bemeneti_ido = input('Adjon meg egy időpontot (óra perc)! ')

bemeneti_ido = bemeneti_ido.split()
bemeneti_perc = 60 * int(bemeneti_ido[0]) + int(bemeneti_ido[1])
```

# 5. Feladat

```python
print('\n5. feladat')

for adat in adatok:
  if adat["vonat"] == bemeneti_vonat:
    if adat["allomas"] == min(allomasok):
      indulasi_perc = adat["ora"] * 60 + adat["perc"]
    if adat["allomas"] == max(allomasok):
      erkezesi_perc = adat["ora"] * 60 + adat["perc"]

menetido = erkezesi_perc - indulasi_perc
keses = menetido - (2 * 60 + 22)

if 0 < keses:
  print(f'A(z) {bemeneti_vonat}. vonat útja {keses} perccel hosszabb volt az előírtnál.')
elif 0 == keses:
  print(f'A(z) {bemeneti_vonat}. vonat útja pontosan az előírt ideig tartott.')
else:
  print(f'A(z) {bemeneti_vonat}. vonat útja {-1 * keses} perccel rövidebb volt az előírtnál.,')
```

# 6. Feladat

```python
print('\n6. feladat')

kimeneti_filenev = f'halad{bemeneti_vonat}.txt'

with open(file=kimeneti_filenev, mode='w') as kimeneti_file:
  for adat in adatok:
    if adat["vonat"] == bemeneti_vonat and adat["indul"] == "E":
      print(f'{adat["allomas"]}. állomás: {adat["ora"]}:{adat["perc"]}', file=kimeneti_file)
```

# 7. Feladat

```python
print('\n7. feladat')
vonat_adatok = dict()

for vonat in vonatok:
  vonat_adatok[vonat] = []

  for allomas in allomasok[1:-1]:
    for adat in adatok:
      if adat["vonat"] == vonat and adat["allomas"] == allomas:
        if adat["indul"] == "E":
          vonat_erkezik = adat["ora"] * 60 + adat["perc"]
        if adat["indul"] == "I":
          vonat_indul = adat["ora"] * 60 + adat["perc"]

    vonat_adatok[vonat].append([vonat_erkezik, vonat_indul])


for vonat in vonatok:
  for adat in adatok:
    if (adat["vonat"] == vonat) and (adat["allomas"] == min(allomasok)):
      vonat_indulas = adat["ora"] * 60 + adat["perc"]

    if (adat["vonat"] == vonat) and (adat["allomas"] == max(allomasok)):
      vonat_erkezes = adat["ora"] * 60 + adat["perc"]

  if vonat_indulas < bemeneti_perc < vonat_erkezes:
    if vonat_adatok[vonat][0][0] > bemeneti_perc:
        print(f"A(z) {vonat}. vonat a 0. és a 1. állomás között járt.")
    for i in range(len(vonat_adatok[vonat])):
      if vonat_adatok[vonat][i][0] <= bemeneti_perc <= vonat_adatok[vonat][i][1]:
        print(f"A(z) {vonat}. vonat a {i + 1}. állomáson állt.")
      if vonat_adatok[vonat][i][1] < bemeneti_perc < vonat_adatok[vonat][i + 1][0]:
        print(f"A(z) {vonat}. vonat a {i + 1}. és a {i + 2}. állomás között járt.")
```
