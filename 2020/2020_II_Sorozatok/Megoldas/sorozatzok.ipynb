# 1. Feladat

```python
reszek = list()

with open(file='lista.txt', mode='r') as forras_file:
  for sorszam, sor in enumerate(forras_file):
    sor = sor.strip()
    if sorszam % 5 == 0:
      resz = dict()
      resz['datum'] = sor
    elif sorszam % 5 == 1:
      resz['cim'] = sor
    elif sorszam % 5 == 2:
      resz['epizod'] = sor
    elif sorszam % 5 == 3:
      resz['hossz'] = sor
    elif sorszam % 5 == 4:
      resz['nezve'] = sor
      reszek.append(resz)
```

# 2. Feladat

```python
print('\n2. feladat')

datumok_szama = 0
for resz in reszek:
  if resz['datum'] != 'NI':
    datumok_szama += 1
print(f'A listában {datumok_szama} db vetítési dátummal rendelkező epizód van.')
```

# 3. Feladat

```python
print('\n3. feladat')

db_nezve = 0
for resz in reszek:
  if resz['nezve'] == '1':
    db_nezve += 1
nezve_arany = db_nezve/len(reszek)

print(f'A listában lévő epizódok {round(100*nezve_arany, 2)}%-át látta.')
```

# 4. Feladat

```python
print('\n4. feladat')

eltoltott_ido = 0
for resz in reszek:
  if resz['nezve'] == '1':
    eltoltott_ido += int(resz['hossz'])

percek = eltoltott_ido % 60
orak = (eltoltott_ido // 60) % 24
napok = eltoltott_ido // (60 * 24)

kimeneti_szoveg = 'Sorozatnézéssel '
if 0 < napok:
  kimeneti_szoveg += f'{napok} napot '
if 0 < orak:
  kimeneti_szoveg += f'{orak} órát '
if 0 < percek:
  kimeneti_szoveg += f'{percek} percet '
kimeneti_szoveg += 'töltött.'

print(kimeneti_szoveg)
```

# 6. Feladat

```python
def hetnapja(ev, ho, nap):
  napok = "v h k sze cs p szo".split()
  honapok = (0, 3, 2, 5, 0, 3, 5, 1, 4, 6, 2, 4)

  if ho < 3: 
    ev -= 1

  return napok[(ev + ev // 4 - ev // 100 + ev // 400 + honapok[ho - 1] + nap) % 7]
```
