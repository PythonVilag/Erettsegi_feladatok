# 1. Feladat

```python
taviratok = list()

with open(file='tavirathu13.txt', mode='r') as forras_file:
  for sor in forras_file:
    sor = sor.strip().split()

    tavirat = dict()
    tavirat["varos"] = sor[0]
    tavirat["ido"] = sor[1]
    tavirat["szel"] = sor[2]
    tavirat["ho"] = int(sor[3])

    taviratok.append(tavirat)
```

# 2. Feladat

```python
print('2. feladat')

bemeneti_varos = input('Adja meg egy település kódját! Település: ')
utolso_ido = taviratok[0]["ido"]

for tavirat in taviratok:
  if tavirat["varos"] == bemeneti_varos:
    if utolso_ido <= tavirat["ido"]:
      utolso_ido = tavirat["ido"]

formazott_ido = utolso_ido[:2]+':'+utolso_ido[2:]
print(f'Az utolsó mérési adat a megadott településről {formazott_ido}-kor érkezett.')
```

# 3. Feladat

```python
print('\n3. feladat')

min_tavirat = taviratok[0]
max_tavirat = taviratok[0]

for tavirat in taviratok:
  if tavirat["ho"] < min_tavirat["ho"]:
    min_tavirat = tavirat
  elif tavirat["ho"] > max_tavirat["ho"]:
    max_tavirat = tavirat

min_telepules = min_tavirat["varos"]
min_idopont = min_tavirat["ido"][:2] + ':' + min_tavirat["ido"][2:]
min_homerseklet = min_tavirat["ho"]

max_telepules = max_tavirat["varos"]
max_idopont = max_tavirat["ido"][:2] + ':' + max_tavirat["ido"][2:]
max_homerseklet = max_tavirat["ho"]

print(f'A legalacsonyabb hőmérséklet: {min_telepules} {min_idopont} {min_homerseklet} fok.')
print(f'A legmagasabb hőmérséklet: {max_telepules} {max_idopont} {max_homerseklet} fok.')
```

# 4. Feladat

```python
print('\n4. feladat')

szelcsend_tavoratok = list()

for tavirat in taviratok:
  if tavirat["szel"] == '00000':
    szelcsend_tavoratok.append(tavirat)

if len(szelcsend_tavoratok) == 0:
  print('Nem volt szelcsent a meresek idejen')
else:
  for szelcsend_tavirat in szelcsend_tavoratok:
    szelcsend_telepules = szelcsend_tavirat["varos"]
    szelcsend_idopont = szelcsend_tavirat["ido"]
    formazott_ido = szelcsend_idopont[:2] + ':' + szelcsend_idopont[2:]
    print(szelcsend_telepules, formazott_ido)
```

# 5. Feladat

```python
print('\n5. feladat')

telepulesnevek = list()
for tavirat in taviratok:
  if tavirat["varos"] not in telepulesnevek:
    telepulesnevek.append(tavirat["varos"])

erteksorok = {nev:list() for nev in telepulesnevek}  #dict, gyakorlatilag lista is lehetne de igy közvetlenebbül lehet utalni rajuk
kph_flagek = {nev:[False, False, False, False] for nev in telepulesnevek}
alacsony = {nev:None for nev in telepulesnevek}
magas = {nev:None for nev in telepulesnevek}
for tavirat in taviratok:
  nev = tavirat[0]
  ora = tavirat[1][:2]
  hom = int(tavirat[3]) # mivel ezzel szamolni kell majd, ezert itt atvaltom intbe
  #5/a feladat
  if ora in {'01', '07', '13', '19'}:
    erteksorok[nev].append(hom)
    if ora == '01':
      kph_flagek[nev][0] = True
    if ora == '07':
      kph_flagek[nev][1] = True
    if ora == '13':
      kph_flagek[nev][2] = True
    if ora == '19':
      kph_flagek[nev][3] = True
  #5/b feladat
  if alacsony[nev] is None or hom < alacsony[nev]:  #az adott varos erso erteke mindenkepp beirodik minimumnak
    alacsony[nev] = hom
  if magas[nev] is None or magas[nev] < hom:  #az adott varos erso erteke mindenkepp beirodik maximumnak
    magas[nev] = hom
for nev in telepulesnevek:
  #5/a feladat
  print(nev, end=' ')
  if all(kph_flagek[nev]):
    erteksor = erteksorok[nev]
    atlag = sum(erteksor)/len(erteksor)
    print(f'Kozephomerseklet: {round(atlag)};', end=' ')
  else:
    print('NA;', end=' ')

  #5/b feladat
  ingadozas = magas[nev]-alacsony[nev]
  print(f'Homerseklet-ingadozat: {ingadozas}')
```

# 6. Feladat

```python
print('\n6. feladat')

for telepulesnev in telepulesnevek:
  kimeneti_filenev = f"{telepulesnev}.txt"

  with open(file=kimeneti_filenev, mode='w') as kimeneti_file:
    print(telepulesnev, file=kimeneti_file)

    for tavirat in taviratok:
      if tavirat["varos"] == telepulesnev:
        ido = tavirat["ido"]
        formazott_ido = ido[:2] + ':' + ido[2:]
        szelerosseg = int(tavirat["szel"][3:])
        print(formazott_ido, '#' * szelerosseg, file=kimeneti_file)
```
