# 1. Feladat

````python
taviratok = list()

with open(file='tavirathu13.txt', mode='r') as forras_file:
  for sor in forras_file:
    tavirat = sor.strip().split()
    taviratok.append(tavirat)
````

# 2. Feladat

````python
print('2. feladat')
bemeneti_varos = input('Adja meg egy település kódját! Település: ')
utolso_ido = '0000'

for tavirat in taviratok:
  if tavirat[0] == bemeneti_varos:
    if utolso_ido <= tavirat[1]:
      utolso_ido = tavirat[1]

formazott_ido = utolso_ido[:2]+':'+utolso_ido[2:]
print(f'Az utolsó mérési adat a megadott településről {formazott_ido}-kor érkezett.')
````

# 3. Feladat

````python
print('\n3. feladat')

min_tavirat = taviratok[0]
max_tavirat = taviratok[0]

for tavirat in taviratok:
  if tavirat[3] < min_tavirat[3]:
    min_tavirat = tavirat
  elif tavirat[3] > max_tavirat[3]:
    max_tavirat = tavirat

min_telepules = min_tavirat[0]
min_idopont = min_tavirat[1][:2]+':'+min_tavirat[1][2:]
min_homerseklet = min_tavirat[3]

max_telepules = max_tavirat[0]
max_idopont = max_tavirat[1][:2]+':'+max_tavirat[1][2:]
max_homerseklet = max_tavirat[3]

print(f'A legalacsonyabb hőmérséklet: {min_telepules} {min_idopont} {min_homerseklet} fok.')
print(f'A legmagasabb hőmérséklet: {max_telepules} {max_idopont} {max_homerseklet} fok.')
````

# 4. Feladat

````python
print('\n4. feladat')

csendes_taviratok = list()

for tavirat in taviratok:
  if tavirat[2] == '00000':
    csendes_taviratok.append(tavirat)

if len(csendes_taviratok) == 0:
  print('Nem volt szelcsent a meresek idejen')
else:
  for csendes_tavirat in csendes_taviratok:
    csendes_telepules = csendes_tavirat[0]
    csendes_idopont = csendes_tavirat[1]
    formazott_ido = csendes_idopont[:2]+':'+csendes_idopont[2:]
    print(csendes_telepules, formazott_ido)
````

# 5. Feladat

````python
print('\n5. feladat')

telepulesnevek = {taviratok[0] for taviratok in taviratok}  #set, a többször elofordulo kodokat osszevonja
erteksorok = {nev:list() for nev in telepulesnevek}  #dict, gyakorlatilag lista is lehetne de igy közvetlenebbül lehet utalni rajuk
kph_flagek = {nev:[False, False, False, False] for nev in telepulesnevek}
alacsony = {nev:None for nev in telepulesnevek}
magas = {nev:None for nev in telepulesnevek}
for tavirat in taviratok:
  nev = tavirat[0]
  ora = tavirat[1][:2]
  hom = int(tavirat[3]) # mivel ezzel szamolni kell majd, ezert itt atvaltom intbe
  #5/a feladat
  if ora in {'01', '07', '13', '19'}:
    erteksorok[nev].append(hom)
    if ora == '01':
      kph_flagek[nev][0] = True
    if ora == '07':
      kph_flagek[nev][1] = True
    if ora == '13':
      kph_flagek[nev][2] = True
    if ora == '19':
      kph_flagek[nev][3] = True
  #5/b feladat
  if alacsony[nev] is None or hom < alacsony[nev]:  #az adott varos erso erteke mindenkepp beirodik minimumnak
    alacsony[nev] = hom
  if magas[nev] is None or magas[nev] < hom:  #az adott varos erso erteke mindenkepp beirodik maximumnak
    magas[nev] = hom
for nev in telepulesnevek:
  #5/a feladat
  print(nev, end=' ')
  if all(kph_flagek[nev]):
    erteksor = erteksorok[nev]
    atlag = sum(erteksor)/len(erteksor)
    print(f'Kozephomerseklet: {round(atlag)};', end=' ')
  else:
    print('NA;', end=' ')

  #5/b feladat
  ingadozas = magas[nev]-alacsony[nev]
  print(f'Homerseklet-ingadozat: {ingadozas}')
````

A feladat szerint ha egy város adatai a következők:
```
0100 12
0115 15

0715 21
0730 24

1300 32

1900 26
```

Akkor a napi középhőmérséklet (12+15+21+24+32+26) / 6, nem pedig ((12+15)/2 + (21+24)/2 + 32 + 26) / 4. 


Ennek semmi értelme, mert ha hajnalban gyakrabban mérünk, akkor eltorzítjuk a hőmérsékletet, de ezt mondja a feladat, úgyhogy ezt kell csinálni.

# 6. Feladat

````python
print('\n6. feladat')

for nev in telepulesnevek:
  kimeneti_filenev = nev+'.txt'
  with open(kimeneti_filenev, 'w') as kimeneti_file:
    print(nev, file=kimeneti_file)
    for tavirat in taviratok:
      if tavirat[0] == nev:
        ido = tavirat[1]
        formazott_ido = ido[:2]+':'+ido[2:]
        szelerosseg = int(tavirat[2][3:])
        print(formazott_ido, '#'*szelerosseg, file=kimeneti_file)
````
