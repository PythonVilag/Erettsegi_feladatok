# 1. Feladat

```python
fogasok = {"leves": [], "foetel": [], "desszert": []}

with open(file='etelek.txt', mode='r', encoding='utf-8') as forras_file:
  for sor in forras_file:
    sor = sor.strip()
    etel = sor.split(" ")[1]
    if "leves" in sor:
      fogasok["leves"].append(etel)
    elif "foetel" in sor:
      fogasok["foetel"].append(etel)
    elif "desszert" in sor:
      fogasok["desszert"].append(etel)

etelek = {}
for fogas in fogasok:
  for etel in fogasok[fogas]:
    with open(file=f"{etel}.txt", mode="r", encoding="utf-8") as forras_file:
      etelek[etel] = forras_file.read()


for fogas in fogasok:
  print(f"{fogas} ({len(fogasok[fogas])}):")
  print(" - ".join(sorted(fogasok[fogas])))
  print()
```

# 2. Feladat

```python
print("2. feladat.")

lisztes_etelek = []

for etel in etelek:
  if "liszt" in etelek[etel]:
    lisztes_etelek.append(etel)

print(f"Lisztet tartalmazó ételek: {', '.join(lisztes_etelek)}")
```

# 3. Feladat

```python
print('3. feladat')

legkevesbe_szeretett_etel = input(f"Az alábbi ételek közül melyiket szereti legkevésbé: {' '.join(lisztes_etelek)}?")

import random
lisztes_etelek.remove(legkevesbe_szeretett_etel)
veletlenszeruen_valasztott_etel = random.choice(lisztes_etelek)

print(f"Véletlenszerűen választott étel: {veletlenszeruen_valasztott_etel}")
```
