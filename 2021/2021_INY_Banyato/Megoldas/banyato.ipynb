# 1. Feladat

````python
melysegek = list()

with open(file='melyseg.txt', mode='r') as forras_file:
  for sorszam, sor in enumerate(forras_file):
    if sorszam == 0:
      sorok_szama = int(sor)
    elif sorszam == 1:
      oszlopok_szama = int(sor)
    else:
      melyseg = sor.strip().split()
      melyseg = list(map(int, melyseg))
      melysegek.append(melyseg)
````

# 2. Feladat

````python
print('2. feladat')

bemeneti_sor = int(input('A m√©r√©s sor√°nak azonos√≠t√≥ja=')) - 1
bemeneti_oszlop = int(input('A m√©r√©s oszlop√°nak azonos√≠t√≥ja=')) - 1

cella_melyseg = melysegek[bemeneti_sor][bemeneti_oszlop]
print(f'A m√©rt m√©lys√©g az adott helyen {cella_melyseg} dm')
````

# 3. Feladat

````python
print('\n3. feladat')

nem_felszin = 0
melyseg_osszeg = 0
for sor in melysegek:
  nem_felszin += sor.count(0)
  melyseg_osszeg += sum(sor)

meresek_szama = sorok_szama * oszlopok_szama
felszin = meresek_szama - nem_felszin
atlag_melyseg = melyseg_osszeg / felszin
print(f'A t√≥ felsz√≠ne: {felszin} m2, √°tlagos m√©lys√©ge: {round(atlag_melyseg/10, 2)} m')
````

# 4. Feladat

````python
print('\n4. feladat')

max_ertek = -1
max_helyek = list()
for sorszam, sor in enumerate(melysegek):
  for oszlopszam, ertek in enumerate(sor):
    if max_ertek < ertek:
      max_ertek = ertek
      max_helyek = [(sorszam + 1, oszlopszam + 1)]
    elif max_ertek == ertek:
        max_helyek.append((sorszam + 1, oszlopszam + 1))

print(f'A t√≥ legnagyobb m√©lys√©ge: {max_ertek} dm')
print('A legm√©lyebb helyek sor-oszlop koordin√°t√°i:')
for max_hely in max_helyek:
  print(f'({max_hely[0]}; {max_hely[1]})', end='\t')
````

# 5. Feladat

````python
print('\n5. feladat')

partszakasz = 0

for sorszam in range(sorok_szama):
  for oszlopszam in range(oszlopok_szama):
    if oszlopszam > 0:
      bal = melysegek[sorszam][oszlopszam-1]
      jobb = melysegek[sorszam][oszlopszam]
      if (bal == 0) != (jobb == 0):
        partszakasz += 1
    if sorszam > 0:
      felso = melysegek[sorszam-1][oszlopszam]
      also = melysegek[sorszam][oszlopszam]
      if (felso == 0) != (also == 0):
        partszakasz += 1

print(f'A t√≥ partvonala {partszakasz} m hossz√∫')
````

# 6. Feladat

````python
print('\n6. feladat')

oszlop_azonosito = int(input('A vizsg√°lt szelv√©ny oszlop√°nak azonos√≠t√≥ja=')) - 1
vizsgalt_oszlop = [melysegek[k][oszlop_azonosito] for k in range(sorok_szama)]

with open(file='diagram.txt', mode='w') as kimeneti_file:
  for sorszam, ertek in enumerate(vizsgalt_oszlop):
    sorszam = f'{sorszam+1:02d}'
    csillagok = round(ertek/10)*'*'
    print(f'{sorszam}{csillagok}', file=kimeneti_file)
````
