# 1. Feladat

````python
# print('1. feladat')

with open('melyseg.txt', 'r') as forras_file:
  melysegek = list()
  for sorszam, sor in enumerate(forras_file):
    if sorszam == 0:
      sorok_szama = int(sor)
    elif sorszam == 1:
      oszlopok_szama = int(sor)
    else:
      melyseg = sor.strip().split()
      melyseg = list(map(int, melyseg))
      melysegek.append(melyseg)
````

# 2. Feladat

````python
print('2. feladat')
bemeneti_sor = int(input('A meres soranak azonositoja=')) - 1
bemeneti_oszlop = int(input('A meres oszlopanak azonositoja=')) - 1

cella_melyseg = melysegek[bemeneti_sor][bemeneti_oszlop]
print(f'A mert melyseg az adott helyen {cella_melyseg} dm')
````

# 3. Feladat

````python
print('\n3. feladat')

nem_felszin = 0
melyseg_osszeg = 0
for sor in melysegek:
  nem_felszin += sor.count(0)
  melyseg_osszeg += sum(sor)

meresek_szama = sorok_szama * oszlopok_szama
felszin = meresek_szama - nem_felszin
atlag_melyseg = melyseg_osszeg / felszin
print(f'A to felszine {felszin} m2, atlagos melysege: {round(atlag_melyseg/10, 2)} m')
````

A felulet a nem-nulla ertekek szama [m^2]-ben az atlag melyseg a melysegek osszege osztva a melysegek szamaval, átváltva m-be

Alternativ:

# 4. Feladat

````python
print('\n4. feladat')

max_ertek = -1
max_helyek = list()
for sorszam, sor in enumerate(melysegek):
  for oszlopszam, ertek in enumerate(sor):
    if max_ertek < ertek:
      max_ertek = ertek
      max_helyek = [(sorszam + 1, oszlopszam + 1),]  #+1 hogy kesobb ne kelljen meg1x vegigmenni rajtuk egyesevel a kiirashoz
    elif max_ertek == ertek:
        max_helyek.append((sorszam + 1, oszlopszam + 1))
print(f'A to legnagyobb melysege: {max_ertek} dm')
print('A legmelyebb helyek sor-oszlop koordinátái:')

print('\t'.join(map(str, max_helyek)))  #hasonlit a peldara, de a sima print is jo
````

# 5. Feladat

````python
print('\n5. feladat')
partszakasz = 0
for sorszam in range(sorok_szama):
  for oszlopszam in range(oszlopok_szama):
    if oszlopszam > 0:
      bal = melysegek[sorszam][oszlopszam-1]
      jobb = melysegek[sorszam][oszlopszam]
      if (bal == 0) != (jobb == 0):
        partszakasz += 1
    if sorszam > 0:
      felso = melysegek[sorszam-1][oszlopszam]
      also = melysegek[sorszam][oszlopszam]
      if (felso == 0) != (also == 0):
        partszakasz += 1

print(f'A tó partvonala {partszakasz} m hosszú')
````

# 6. Feladat

````python
print('\n6. feladat')
oszlop_azonosito_int = int(input('A vizsgalt szelveny oszlopanak azonositoja=')) - 1
with open('diagram.txt', 'w') as kimeneti_file:
  vizsgalt_oszlop = [melysegek[k][oszlop_azonosito_int] for k in range(sorok_szama)]
  for sorszam, ertek in enumerate(vizsgalt_oszlop):
    print(f'{sorszam:02d}{"*"*round(ertek/10)}', file=kimeneti_file)
````
