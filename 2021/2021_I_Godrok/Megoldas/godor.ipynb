# 1. Feladat

```python
print('1. feladat')

melysegek = list()
with open(file='melyseg.txt', mode='r') as forras_file:
  for sor in forras_file:
    elem = int(sor)
    melysegek.append(elem)

sorok_szama = len(melysegek)
print(f'A fájl adatainak száma: {sorok_szama}')
```

# 2. Feladat

```python
print('\n2. feladat')

bemeneti_sorszam = int(input('Adjon meg egy távolságértéket! ')) - 1
print(f'Ezen a helyen a felszín {melysegek[bemeneti_sorszam]} méter mélyen van.')
```

# 3. Feladat

```python
print('\n3. feladat')

nullak_szama = melysegek.count(0)
nullak_aranya = nullak_szama/len(melysegek)
print(f'Az érintetlen terület aránya {round(100*nullak_aranya, 2)}%.')
```

# 4. & 5. Feladat

```python
godrok_szama = 0
with open(file='godrok.txt', mode='w') as kimeneti_file:
  godor_lista = list()
  for ertek in melysegek:
    if ertek == 0:
      if 0 < len(godor_lista):
        print(' '.join(godor_lista), file=kimeneti_file)
        godor_lista = list()
        godrok_szama += 1
    else:
      godor_lista.append(str(ertek))

print('\n5. feladat')
print(f'A gödrök száma: {godrok_szama}')
```

# 6. Feladat

```python
print('\n6. feladat')

if melysegek[bemeneti_sorszam] == 0:
  print('Az adott helyen nincs gödör.')
else:
  print('a)')

  kezdet = bemeneti_sorszam
  while melysegek[kezdet] != 0:
    kezdet -= 1
  veg = bemeneti_sorszam
  while melysegek[veg] != 0:
    veg += 1

  kezdet += 1
  veg -= 1
  print(f'A gödör kezdete: {kezdet+1} méter, a gödör vége: {veg+1} méter.')
    
    
  print('b)')
  pozicio = kezdet

  while melysegek[pozicio] >= melysegek[pozicio - 1] and pozicio <= veg:
    pozicio += 1
  while melysegek[pozicio] <= melysegek[pozicio - 1] and pozicio <= veg:
    pozicio += 1
  if pozicio == veg + 1:
    print('Folyamatosan mélyül.')
  else:
    print('Nem mélyül folyamatosan.')
    
    
  print('c)')
  print(f'A legnagyobb mélysége {max(melysegek[kezdet:veg+1])} méter.')

  print('d)')
  melyseg_osszeg = sum(melysegek[kezdet:veg+1])
  print(f'A térfogata {melyseg_osszeg*10} m^3.')
  
  print('e)')
  viz_osszeg = melyseg_osszeg - (veg-kezdet+1)
  print(f'A vízmennyiség {10*viz_osszeg} m^3.')
```
